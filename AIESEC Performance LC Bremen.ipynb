```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Period': ['2019-2020', '2020-2021', '2021-2022', '2022-2023', '2023-2024', '2024-2025'],
    '#SUs': [158, 112, 219, 294, 149, 376],
    '#APP': [22, 8, 12, 7, 5, 8],
    '#ACCH': [2, 0, 0, 1, 0, 3],
    '#APD': [0, 0, 0, 1, 1, 2],
    '#REA': [0, 0, 0, 0, 0, 0],
    '#FIN': [3, 1, 3, 2, 4, 0],
    '#CO': [2, 0, 1, 0, 0, 2]
}
df = pd.DataFrame(data)
df.set_index('Period', inplace=True)
metrics = ['#SUs', '#APP', '#ACCH', '#APD', '#REA', '#FIN', '#CO']
for metric in metrics:
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df[metric], marker='o', label=metric)
    plt.title(f'oGV Performance: {metric} from 2019-2020 to 2024-2025')
    plt.xlabel('Period')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
df.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('oGV Overall Performance by Period')
plt.xlabel('Period')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.tight_layout()
plt.show()
```
